## <span style='color:blue'>2. Business Questions

### Q1. Find the total number of products for each product category

```python
%%sql
SELECT 
    PRODCATEGORYID, COUNT(PRODUCTID) AS total_product
FROM 
    products
GROUP BY
    PRODCATEGORYID;
```

### Q2. List the top 5 most expensive products

```python
%%sql
SELECT 
    PRODUCTID, PRODCATEGORYID, PRICE
FROM 
    products
ORDER BY 
    PRICE DESC LIMIT 5;
```

### Q3: Find all products that belong to the 'Mountain Bike' category

```python
%%sql
SELECT 
    PRODUCTID, SHORT_DESCR
FROM 
    products P LEFT OUTER JOIN
    prodcattext T on
    P.PRODCATEGORYID = T.PRODCATEGORYID
WHERE
    SHORT_DESCR LIKE '%MOUNTAIN BIKE%';
```

### Q4: List the total sales amount (gross) for each product category

```python
%%sql
SELECT 
    PRODCATEGORYID, round(SUM(GROSSAMOUNT),0) AS Gross_Sales
FROM 
    orders O LEFT OUTER JOIN products P
    ON O.PARTNERID = P.SUPPLIER_PARTNERID
GROUP BY 
    PRODCATEGORYID
ORDER BY 
    Gross_Sales DESC;    
```

### Q5: List the top 5 suppliers by total product sales.

```python
%%sql
SELECT 
    PARTNERID, SUM(GROSSAMOUNT) AS total_sales 
FROM 
    orders
GROUP BY
    PARTNERID
ORDER BY 
    total_sales DESC
LIMIT 5; 
```

### Q6. Find the total number of products created by each employee

```python
%%sql
SELECT 
    NAME_FIRST, NAME_LAST, COUNT(PRODUCTID) AS total_prod
FROM 
    products P LEFT OUTER JOIN employees E
    ON P.CREATEDBY = E.EMPLOYEEID
GROUP BY 
    NAME_FIRST, NAME_LAST
ORDER BY
    total_prod;
    
```

### Q7. List the employees who have changed product details the most

```python
%%sql
SELECT 
    NAME_FIRST, NAME_LAST, COUNT(CHANGEDBY) AS change_count
FROM
    products P LEFT OUTER JOIN employees E
    ON P.CREATEDBY = E.EMPLOYEEID
GROUP BY
    NAME_FIRST, NAME_LAST
ORDER BY
    change_count DESC
LIMIT 5;
```

### Q8. Calculate the total gross amount for each sales order.

```python
%%sql
SELECT 
    SALESORDERID, GROSSAMOUNT
FROM 
    orders
ORDER BY 
    GROSSAMOUNT DESC
LIMIT 10;
```

### Q9. Trend in sales over different fiscal year periods

```python
%%sql
SELECT 
    SUBSTRING(FISCALYEARPERIOD, 1, 4) AS year, round(AVG(GROSSAMOUNT),0) AS avg_gross_sales
FROM 
    orders
GROUP BY 
    SUBSTRING(FISCALYEARPERIOD, 1, 4)
ORDER BY
    avg_gross_sales;
```

### Q10. Which products contribute the most to revenue when the billing status is 'Complete'

```python
%%sql
SELECT 
    PRODUCTID, round(AVG(GROSSAMOUNT),0) AS avg_gross_sales
FROM 
    orders O LEFT OUTER JOIN Products P
    ON O.CREATEDBY = P.CREATEDBY
WHERE
    BILLINGSTATUS = 'C'
GROUP BY 
    PRODUCTID
ORDER BY
    avg_gross_sales DESC
LIMIT 5;
    
```

### Q11. Find the sales order items for a specific product ID.

```python
%%sql
SELECT 
    PRODUCTID, COUNT(SALESORDERITEM) AS order_items
FROM 
    items
GROUP BY
    PRODUCTID
LIMIT 5
    ;
```

### Q12: How does revenue vary over different fiscal year periods for orders with a billing status of 'Complete'

```python
%%sql
SELECT 
    SUBSTRING(FISCALYEARPERIOD, 1, 4) AS year, round(AVG(GROSSAMOUNT),0) AS avg_gross_sales
FROM 
    orders
WHERE 
    BILLINGSTATUS = 'C'
GROUP BY 
    SUBSTRING(FISCALYEARPERIOD, 1, 4)
ORDER BY
    avg_gross_sales;
```

### Q13. Find the top-selling product within each category along with its total sales amount.

```python
%%sql
SELECT 
    PRODCATEGORYID, PRODUCTID, round(SUM(GROSSAMOUNT)) AS total_sales
FROM 
    orders O 
    LEFT OUTER JOIN products P
    ON O.CREATEDBY = P.CREATEDBY
GROUP BY 
        PRODCATEGORYID, PRODUCTID
ORDER BY PRODCATEGORYID , total_sales DESC
;
```

### Q14. Calculate the total gross amount for each sales organization.

```python
%%sql
SELECT 
    SALESORG, round(SUM(GROSSAMOUNT)) AS total_sales
FROM 
    orders
GROUP BY 
    SALESORG
ORDER BY
    total_sales DESC;
```

### Q15: Find the top 5 sales orders by net amount.

```python
%%sql
SELECT 
    SALESORDERID, round(SUM(NETAMOUNT), 0) AS net_total
FROM 
    orders O 
    LEFT OUTER JOIN Products P
    ON O.CREATEDBY = P.CREATEDBY
GROUP BY 
    SALESORDERID
ORDER BY 
    net_total DESC
LIMIT 5;
```

### Q16. How many sales orders were created in the year 2018?

```python
%%sql
SELECT 
    SUBSTRING(FISCALYEARPERIOD, 1, 4) AS Year, COUNT(SALESORDERID) AS sales_count
FROM 
    orders
GROUP BY 
    Year
ORDER BY
    Year;
```

### Q17: Find the number of employees for each sex.

### Q18. List the employees who have 'W' in their first name

### Q19. Count the number of addresses in each country.

```python
%%sql
SELECT 
    COUNTRY, COUNT(ADDRESSID) AS adress_count
FROM
    addresses
GROUP BY 
    COUNTRY
ORDER BY
    adress_count DESC;
```

### Q20. Find the number of business partners in each region.

```python
%%sql
SELECT REGION, COUNT(*) AS partner_num
FROM 
    partners P LEFT OUTER JOIN addresses A
    ON A.ADDRESSID = P.ADDRESSID
GROUP BY 
    REGION
ORDER BY partner_num DESC;
```
